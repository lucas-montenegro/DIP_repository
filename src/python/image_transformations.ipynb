```python
import cv2 
import numpy as np
import math 
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Crop and flip an image using Numpy array indexing.
#Reading an image
img = cv2.imread("./lena.jpg")
#cv2 reads it in bgr, so we change it to rgb
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#getting a numpy array from the image
arr = np.array(img)
print(arr.shape)
#cropping the image
arr1 = arr[100:200, 0:125]
#Show the image with matplotlib
plt.imshow(arr1)
plt.show()
#flipping the image
arr2 = np.flipud(arr)
#Show the image with matplotlib
plt.imshow(arr2)
plt.show()
```

```python
#Implement image translation using Numpy and OpenCV.
#Reading an image
img = cv2.imread("./lena.jpg")
#cv2 reads it in bgr, so we change it to rgb
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#getting a numpy array from the image
arr = np.array(img)
print(arr.shape)

#translating an image with numpy 
arr1 = np.roll(arr, 30, (0, 1))
#inserting black because np.roll acts like an circle array
arr1[:30] = 0
arr1[:, :30] = 0
#Show the image with matplotlib
plt.imshow(arr1)
plt.show()

#translating an image with opencv
height, width = img.shape[:2]
tx = width / 10
ty = height / 6
# creating the translation matrix using tx and ty
translation_matrix = np.array([[1.0, 0.0, tx], [0.0, 1.0, ty]])
translated_img = cv2.warpAffine(src=img, M=translation_matrix, dsize=(width, height))
#Show the image with matplotlib
plt.imshow(translated_img)
plt.show()
```

```python
#Implement image rotation using Numpy and OpenCV.
#Reading an image
img = cv2.imread("./lena.jpg")
#cv2 reads it in bgr, so we change it to rgb
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#getting a numpy array from the image
arr = np.array(img)
print(arr.shape)

#rotating an image with numpy (only 90, 180, 270 or 360 degrees)
arr1 = np.rot90(arr, 1)
#Show the image with matplotlib
plt.imshow(arr1)
plt.show()

#rotating an image with opencv
height, width = img.shape[:2]
center = (width/2, height/2)
rotate_matrix = cv2.getRotationMatrix2D(center=center, angle=60, scale=1)
rotated_img = cv2.warpAffine(src=img, M=rotate_matrix, dsize=(width, height))
#Show the image with matplotlib
plt.imshow(rotated_img)
plt.show()
```

```python
#Implement image resizing using OpenCV.
#Reading an image
img = cv2.imread("./lena.jpg")
#cv2 reads it in bgr, so we change it to rgb
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
height, width = img.shape[:2]

#resizing the image
new_size = (int((width/100) * 60), int((height/100) * 45))
img = cv2.resize(img, dsize=new_size)

#Show the image with matplotlib
plt.imshow(img)
plt.show()
```

```python
#Implement bitwise operations: AND, OR, XOR.
#drawing a rectangle
rectangle = np.zeros((200, 200, 3))
rectangle = cv2.rectangle(rectangle, (20, 20), (180, 180), (255, 255, 255), -1)

#drawing a circle
circle = np.zeros((200, 200, 3))
cv2.circle(circle, (100, 100), 100, (255, 255, 255), -1)

#AND operation
bitwise_and = cv2.bitwise_and(rectangle, circle)
plt.imshow(bitwise_and)
plt.show()

#OR operation
bitwise_or = cv2.bitwise_or(rectangle, circle)
plt.imshow(bitwise_or)
plt.show()

#XOR operation
bitwise_xor = cv2.bitwise_xor(rectangle, circle)
plt.imshow(bitwise_xor)
plt.show()
```

```python
#Implement the "mask" operation, where a third image 'h' contains only a Region of Interest (ROI -- defined by the second image mask 'g') obtained from the input image 'f'. Note that this Region can be of any shape.

#Reading an image
img = cv2.imread("./lena.jpg")
#cv2 reads it in bgr, so we change it to rgb
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

circle_mask = np.zeros(img.shape[:2], dtype="uint8")
cv2.circle(circle_mask, (70, 95), 50, 255, -1)
masked_img = cv2.bitwise_and(img, img, mask=circle_mask)

#Show the image with matplotlib
plt.imshow(masked_img)
plt.show()
```
