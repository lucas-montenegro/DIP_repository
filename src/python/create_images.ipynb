```python
import cv2 
import numpy as np
import math 
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Create the image of a paraboloid with one axis scaled (like an oval paraboloid).
x = np.linspace(-50, 50, 50)
y = np.linspace(-100, 100, 200)

x, y = np.meshgrid(x, y)
a = 1.5
b = 3
c = 3

x_axis = (x**2) / (a**2)
y_axis = (y**2) / (b**2)
z = c*(x_axis + y_axis)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap="plasma", linewidth=0, antialiased=False, alpha=0.5)
plt.show()
```

```python
#Create the image of a rotated sin using rotation of coordinates.
x = np.linspace(-2*np.pi, 2*np.pi, 200)
y = np.linspace(-1, 1, 200)
x, y = np.meshgrid(x, y)

#sin function
a = 1
b = 1
c = 0
d = 0
y = a*np.sin(((b * x) + c)) + d


plt.plot(x, y, color = 'red', marker = "o")
plt.show()

#rotating it 
theta = math.pi / 6
x1 = (x * math.cos(theta)) + (y * math.sin(theta))
y1 = -(x * math.sin(theta)) + (y * math.cos(theta))
plt.plot(x1, y1, color = 'red', marker = "o")
plt.show()
```

```python
#Create the image of a gaussian.
x = np.linspace(-10, 10, 500)
y = np.linspace(-10, 10, 500)
x, y = np.meshgrid(x, y)

#gaussian function
a = 1
b = 1
c = 1
y = a*np.exp(- (((x - b)**2) + (2 * c**2)))

plt.plot(x, y, color = 'red', marker = "o")
plt.show()
```
